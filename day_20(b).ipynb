```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
year = [1950,1970,1990,2010]
pop = [2.5,3.6,5.2,6.9]
plt.plot(year,pop)
plt.show()
```

```python
year = [1950,1970,1990,2010]
pop = [2.5,3.6,5.2,6.9]
plt.scatter(year,pop)
plt.xlabel("Years")
plt.ylabel("Population in Cr")
plt.title("Population Increase in Years")
plt.show()
```
