```python
import numpy as np
import pandas as pd
from random import shuffle
```

```python
marks = np.arange(41,101)
shuffle(marks)
marks = marks.reshape(12,5)
```

```python
name = ['shaina','yogita','divya','aditya','tanishq','harsh','Ana','Nonu','Aman','Nobita','yammi','divi']
subject = ['hindi','english','physics','maths','chemistry']
```

```python
grade = pd.DataFrame(data = marks,index = name,columns = subject)
grade
```

```python
df = grade.loc[:,['physics','maths','chemistry']]
df
```

```python
df['Average'] = df.mean(axis = 1)
```

```python
df
```

```python
df[df['Average'] >= 65]
```

```python
df.to_csv("Selected_candidate")
```
