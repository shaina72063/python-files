```python
years = [1950 , 1960 , 1970 , 1980 , 1990 , 2000 , 2010 ,2020]
company_x = [2.5,3.5,2.8,3.1,4.0,2.9,4.5,4.8]

years = [1950 , 1960 , 1970 , 1980 , 1990 , 2000 , 2010 ,2020]
company_y = [2.4,3.2,2.7,3.5,4.3,4.0,2.4,3.8]

font1 = {'family':'serif','color':'brown','size':14}
font2 = {'family':'serif','color':'brown','size':15}
```

```python
plt.subplot(1,2,1)
plt.plot(years,company_x, color = 'blue' ,marker ='o',ms =10 , mfc ='pink' , mec = 'red')
plt.grid(linestyle = 'dashed',color ='green')
plt.xlabel("years")
plt.ylabel("Company_x Capital")
plt.title("Company_x capital over years",fontdict = font1)

plt.subplot(1,2,2)
plt.plot(years,company_y, color = 'green' ,marker ='D',ms =10 , mfc ='pink' , mec = 'red')
plt.grid(linestyle = 'dashed',color ='blue')
plt.xlabel("years")
plt.ylabel("Company_y Capital")
plt.title("Company_y capital over years",fontdict = font2)
plt.subplots_adjust(wspace = 1)

plt.suptitle("Capital comparision of companies over years")
plt.show()
```
