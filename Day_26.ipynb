```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("world_info.csv")
data.head()
```

```python
df = data.loc[:,['country','population','area','infant_maternity','gdp_percapita','literacy_percent','birthrate','deathrate']]
df.head()
```

```python
df1 = df.sort_values(by = ['population'], ascending = False )
df1.head(10)
```

```python
df2=df1.iloc[0:10]
df2
```

```python
plt.pie(df2['population'] , labels = df2['country'] , startangle = 90)
plt.show()
```

```python
plt.pie(df2['area'] , labels = df2['country'] , startangle = 90)
plt.show()
```
