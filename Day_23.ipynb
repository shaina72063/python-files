```python
df1 = pd.read_csv("student-alcohol-consumption.csv")
df1.head()
```

```python
sns.countplot(x="study_time",data=df1)
plt.show()
```

```python
sns.scatterplot(x="school",y="absences",hue="traveltime",data=df1)
plt.show()
```

```python
df1['school'].unique()
```

```python
sns.countplot(x="age",data=df1,hue="sex")
plt.show()
```

```python
sns.relplot(x="G1",y="absences",data=df1,kind="scatter",col="location",row="sex")
plt.show()
```

```python
sns.relplot(x="G1",y="G3",data=df1,kind="scatter")
plt.show()
```

```python
sns.relplot(x="G1",y="G3",data=df1,kind="scatter",col="schoolsup",col_order = ["yes","no"])
plt.show()
```

```python
sns.relplot(x="G1",y="G3",data=df1,kind="scatter",col="schoolsup",col_order = ["yes","no"],row = "famsup",row_order=["yes","no"])
plt.show()
```
